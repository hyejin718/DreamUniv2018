```python
# 팩토리얼 계산하기
n = int(input("정수 입력 바람: "))
fact = 1
for i in range(1, n+1):
    fact = fact * i
print(n, "팩토리얼은", fact,"이다.")
```

```python
# 팩토리얼 계산하기
n = int(input("정수 입력 바람: "))
fact = 1
for i in range(1, n+1):
    fact = fact * i
    print(fact)  # 계산 진행상황알아보기!!!!
print(n, "팩토리얼은", fact,"이다.")
```

```python
# while 구문 만들기~~~~~~~~~~~~~~~~~~~~~~~~~~
response = "아니"
while response == "아니":
    print("아들은 혼자 놀고 있습니다.")
    response = input("엄마, 다됐어?")
print("밥 먹자아아아")
```

```python
password = "" # 초기값
while password != "패스워드":                         # password가 비밀번호와 같지 않은 상태가 참임
    password = input("암호를 입력하세요: ")
    print("비밀번호가 틀립니다.")
print("로그인 성공")
```
